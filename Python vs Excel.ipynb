# Python Function 1: FILTERING (LOC/ILOC)

```python
import pandas as pd
dog=pd.read_excel("Dog breeds ranked.xlsx")
dog.head()
```

```python
dog[(dog['score'] >3.0) & (dog['intelligence']== "Above average")]
```

```python
dog.loc[(dog['score'] >3.0) & (dog['intelligence']== "Above average"),:]
```

# 2) SORT

```python
dog[(dog['score']>3.0) & (dog['intelligence'] == "Above average")].sort_values('popularity ranking',ascending=False)
```

# 3) Groupby instead of SUMIF

```python
grouped = dog.groupby('type')['score'].sum()
grouped['terrier']
```

# 4) Groupby instead of SUMIF(S)

```python
grouped = dog.groupby(['type','size.1'])['score'].sum()
grouped['terrier']['small']
```
